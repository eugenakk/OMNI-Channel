```python
# 7월~ 12월 데이터프레임을 BARCODE기준으로 JOIN시키기 
#----------> OMNI_GS 데이터프레임에 하나로 저장!!!!

#df2 - 7월 데이터
#df3- 8월 데이터
#df4 - 9월 데이터
# df5- 10월 데이터
# df6 = 11월 데이터
# df7 = 12월 데이터
```

```python
#merge사용
df_merge_1= pd.merge(df2, df3, how='outer', on=['BARCODE','CUSTOM_NAME','Product_name(Customer)'])
```

```python
df_merge_2 = pd.merge(df_merge_1, df4, how='outer', on=['BARCODE','CUSTOM_NAME','Product_name(Customer)'])
```

```python
df_merge_3 = pd.merge(df_merge_2, df5, how='outer', on=['BARCODE','CUSTOM_NAME','Product_name(Customer)'])
```

```python
df_merge_4 = pd.merge(df_merge_3, df6, how='outer', on=['BARCODE','CUSTOM_NAME','Product_name(Customer)'])
```

```python
OMNI_EMART_HM = pd.merge(df_merge_4, df7, how='outer',on=['BARCODE','CUSTOM_NAME','Product_name(Customer)'])
```

```python
#613개의 제품에 대한 "거래"가 있었고, 그것에 대한 데이터프레임
OMNI_EMART_HM
```

```python
# P3M, P6M --->에 대한 GAP을 구하기.
#먼저 NaN을 0으로 처리하기
OMNI_EMART_HM =OMNI_EMART_HM  .fillna(0)

#make new column (P3M gap)
OMNI_EMART_HM  ['P3M GAP']= OMNI_EMART_HM .apply(lambda row: (row[6] + row[7] + row[8])/3 ,axis=1)

#make new column (P6M gap)
OMNI_EMART_HM ['P6M GAP']=OMNI_EMART_HM .apply(lambda row: (row[3] + row[4] + row[5]+row[6] + row[7] + row[8])/6,axis=1)
```
